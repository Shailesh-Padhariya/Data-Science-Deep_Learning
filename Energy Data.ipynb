```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform (df_input)
display (data_scaled)
```

```python
features = data_scaled
display (features)
```

```python
target = data_scaled[:,0]
display (target)
```

```python
pd.DataFrame(data_scaled[:,0])
```

```python
pd.DataFrame(features).head(20)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
TimeseriesGenerator(features, target,length = 2, sampling_rate = 1 , batch_size =1)[0]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features,target, test_size = .2, random_state = 123, shuffle = False)
display (x_train.shape)
display (x_test.shape)
display (y_train.shape)
display (y_test.shape)
```

```python
ws= 720 # use 5 days of data to prediction 6* 24*5 =720
batch_size = 32 # Pass 32 data set at a time
num_features =4
train_generator = TimeseriesGenerator(x_train , y_train, length = ws, sampling_rate = 1 , batch_size =batch_size)
display (train_generator[0]) # This display one batch all the required input for 32 prediction 
```

```python
test_generator = TimeseriesGenerator(x_test, y_test, length = ws, sampling_rate = 1 , batch_size =batch_size)
```

```python
display (test_generator [0])
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import LeakyReLU
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential() 
```

```python
model.add(LSTM (units=128, input_shape=(ws,num_features), return_sequences = True))
```

```python
model.add (LeakyReLU(alpha=0.5))
```

```python
model.add(LSTM (units=128,return_sequences = True))
model.add (LeakyReLU(alpha=0.5))
model.add (Dropout(0.3)) 
```

```python
model.add(LSTM (units=64,return_sequences = False))
model.add (Dropout(0.3))
```

```python
model.add (Dense(1))
```

```python
model.summary()
```

```python
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor = 'val_loss', patience =2, mode = 'min')
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ["MeanAbsoluteError"])
```

```python
history = model.fit(train_generator, epochs = 10, validation_data = test_generator, shuffle = False, callbacks = [early_stopping])
```

```python
model.evaluate(test_generator, verbose = 0)
```

```python
predictions = model.predict (test_generator) 

display (predictions.shape) 
# There might be a difference of 720 records , since initial 720 records used for first predictions 

display (predictions ) # The predictions are scaled predictions
```
